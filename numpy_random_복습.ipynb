## 파이썬 random 함수

```python
# 파이썬 random 함수
import random

random.random() # -.0(포함)부터 1.0(불포함) 사이의 수
random.randint(1,10) # 1~10(포함) 사이의 정수 중에서 난수 
random.uniform(10,20) # min, max 사이의 float
random.randrange(10) # 지정범위의 int 
random.choice([1,2,3,4,5]) # 리스트 내부에 있는 요소를 랜덤하게 선택
li = [1,2,3,4,5]
random.sample(li,5) # 리스트 요소를 중복 안되게 리턴 - 리스트에 반영 x
random.shuffle(li) # 리스트 요소를 다시 섞어서 리턴 - 리스트에 반영
print(li)
```

## numpy의 np.random

```python
# numpy의 np.random
# rand: 0.0 ~ 1.0 사이의 균일 분포
# randn: 가우시안 표준 정규 분호
# randint: 균일 분포의 정수 난수
```

```python
# 0.0 ~ 1.0 사이의 균일 분포인 실수
import numpy as np
np.random.seed(0)
print(np.random.rand(5), '\n')
print(np.random.rand(3,2))
```

```python
print(np.random.randint(5),'\n') # 0~4까지 정수 난수 1개
print(np.random.randint(1,20),'\n') # 1~19까지 정수 난수 1개
print(np.random.randint(10, size=10),'\n') # 0~9 정수 난수 10개
print(np.random.randint(10,20, size=5),'\n') # 10~19 정수 난수 5개
print(np.random.randint(10,20, size=(3,5)),'\n') # 10~19 정수난수 (3,5) 
```

```python
print(np.random.choice([1,2,3,4,5]),'\n') # 하나
print(np.random.choice(5,5,replace=False),'\n') # 0~5 정수 5개 중복 안되게
print(np.random.choice(5,5)) # replace=True, 0~5 정수 5개 중복 허용
```

```python
li = [0,1,2,3,4]
np.random.shuffle(li)
li
```

```python
print(np.random.randn(6),'\n')
print(np.random.randn(3,2))
```

```python
# 범위내 실수 출력
print(np.random.uniform(1,2,3),'\n') # 1~2 사이의 실수 3개
print(np.random.uniform(0,1,(2,3))) # 0~1 사이의 실수 (2,3) 형태로 출력
```

```python
# random_sample은 uniform 하수의 최소값, 최대값을 각각 0,1로 지정한 것과 같음
np.random.random_sample((2,3))
# np.random.uniform(0,1,(2,3)) 이랑 같음
```
