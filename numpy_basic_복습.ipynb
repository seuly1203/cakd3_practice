## numpy 배열 만들기

- zeros(): 해당 배열에 모두 0 
- ones(): 해당 배열에 1 
- full(): 배열에 사용자가 지정한 값을 넣는데 사용

### linspace()
- return evenly spaced numbers over a specified interval.
- linspace(start, stop, num=50(default), endpoint=True, retstep=False, dtype=None, axis=0)
- the endpoint of the interval can optionally be excluded: False 설정
- if retstep is True, return (samples, step), where step is the spacing between samples

```python
np.linspace(0,10) # 0 ~ 10을 균등하게 50개로 나누어줌
```

```python
np.linspace(1,11, num=9, retstep=True) # retstep: 간격
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0,1)
y = np.linspace(0,50)
plt.plot(x,y,'o')
```

```python
x = np.linspace(0,20,20)
y = np.linspace(0.1,1,20)
plt.plot(x,y,'*')
```

## sort()

```python
# np.sort(): 정렬, 원본에 미반영 (inplace=False의 느낌)
import numpy as np
ar = np.random.randn(5)
print(ar, '\n')
print(np.sort(ar), '\n')
print(ar)
```

```python
# 행렬이 2차원 이상일 경우 axis 축 값 설정을 통해 행, 열 방향으로 정렬 수행
ar2 = np.array([[8,12],
                [7,1]])
sort_ar2_a0 = np.sort(ar2, axis=0) # 행방향 정렬 (위->아래)
sort_ar2_a1 = np.sort(ar2, axis=1) # 열방향 정렬 ->
display(sort_ar2_a0, sort_ar2_a1)
```

```python
# ndarray.sort(): 원본 반영 (inplace=True처럼)
ar = np.random.randn(6)
print(ar, '\n')
ar.sort()
print(ar)
```

```python
np.random.seed(0)
ar2 = np.random.randn(5,3)
print(ar2, '\n')
ar2.sort(-1) 
print(ar2, '\n')
ar2.sort(1) # 1이나 -1이나 같음
print(ar2)
```

## 선형대수

벡터 a를 벡터 b와 같은 방향의 성분과 그렇지 않은 성분으로 분해한다고 할 때 $|a|cos𝞱$의 길이는 벡터 b와 같은 방향의 성분 길이와 같다.

벡터 a와 b의 내적은 벡터 b의 길이와 벡터 a에서 벡터 b와 같은 방향의 성분 길이를 곱한 것이다.

행렬 내적은 행렬 곱이며 두 행렬 A와 B의 내적은 np.dot()을 이용해 계산이 가능
행렬 내적의 특성으로 왼쪽 행렬의 열개수와 오른쪽 행렬의 행개수가 동일해야 내적 연산 가능

```python
# 행렬곱: x.dot(y), np.dot(x,y)
x = np.arange(1,7).reshape(2,3) #2*3 행렬
y = np.array([[6,23],[-1,7],[8,9]]) #3*2 행렬
print(x,'\n')
print(y,'\n')
np.dot(x,y)
```

```python
# 역행렬: 어떤 행렬 A와 곱했을 때 곱셈에 대한 항등원인 단위행렬 E가
# 나오는 행렬을 행렬 A의 역행렬이라고 함. 
# np.linalg.inv()

np.random.seed(0)
mt1 = np.random.randint(1,4, size=(3,3))
mt2 = np.linalg.inv(mt1)

display(mt1, mt2, np.dot(mt1, mt2))
```
